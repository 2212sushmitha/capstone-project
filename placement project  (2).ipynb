Scope:
   
   The scope of analyzing salary trends includes examining the impact of job titles, industries, and geographic regions on        compensation levels. 
   It also covers identifying variations in pay based on experience, education, and market demand across different locations.

Business_Problem:
    
    A key business problem is determining how to set competitive salaries across different job titles and locations to 
    attract and retain top talent.
    Failure to align compensation with market trends can lead to employee turnover, talent shortages, and reduced business         performance.

```python
import numpy as np
import pandas as pd
from scipy import stats

# Sample dataset with index
data = {
    'Job Id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Experience': [2, 5, 7, 3, 10, 50, 6, 8, 4, 100],  # Outliers in Experience
    'Salary Range': [40000, 50000, 60000, 45000, 70000, 150000, 55000, 65000, 48000, 250000],  # Outliers in Salary
    'Latitude': [12.9716, 13.0827, 12.2958, 11.0168, 19.0760, 72.8777, 28.7041, 77.1025, 22.5726, 88.3639],  # Normal
    'Longitude': [77.5946, 80.2707, 76.6394, 76.9558, 72.8777, 19.0760, 77.1025, 28.7041, 88.3639, 22.5726]  # Normal
}

df = pd.DataFrame(data)
df.index.name = 'Index'  # Ensure index is named

# Select only numeric columns
numeric_cols = ['Experience', 'Salary Range']

# Compute Z-scores for each column
z_scores = np.abs(stats.zscore(df[numeric_cols]))

# Identify outliers (Z-score > 3)
outliers = df[(z_scores > 3).any(axis=1)]

print("Outliers with Index using Z-score:\n", outliers)
```

```python
#Bar Chart: Number of Jobs per Work Type
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("job_descriptions.csv")  


# Set Style
sns.set(style="whitegrid")

# Bar Chart
plt.figure(figsize=(10, 6))
sns.countplot(x=df['Work Type'], palette="viridis")
plt.title("Number of Jobs per Work Type")
plt.xlabel("Work Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#key insights for Bar chart : 
1.This bar chart shows the distribution of jobs by work type, highlighting which types are most and least common. 
2.It helps in understanding job trends and demand for different work types.
```

```python
#Histogram: Experience Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Experience'], bins=10, kde=True, color="blue")
plt.title("Experience Distribution")
plt.xlabel("Years of Experience")
plt.ylabel("Count")
plt.show()

```

```python
#key insights for histogram chart : 
1.This histogram shows the distribution of job experience levels, revealing how many jobs require different years of experience. 
2.The KDE curve helps visualize trends, such as whether most jobs need low, mid, or high experience.
```

```python
# Line Chart: Salary Trend Over Experience
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Experience'], y=df['Salary Range'], marker="o", color="red")
plt.title("Salary Trend Over Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary Range")
plt.grid()
plt.show()
```

```python
#key insights for line chart :
1.This line chart shows how salary changes with experience, helping to identify salary growth trends.
2.The markers highlight specific data points, making it easier to see patterns and anomalies.
```

```python
#Job Posting Trends Over Time (Time Series Plot)

df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])

# Group by Date and Count Jobs
job_trends = df.groupby(df['Job Posting Date'].dt.date).size()

plt.figure(figsize=(12, 6))
plt.plot(job_trends.index, job_trends.values, marker="o", linestyle="-", color="green")
plt.title("Job Posting Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Job Postings")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#key insights for time series plot:
1.This time series plot shows job posting trends over time, highlighting peaks and dips in job availability.
2.It helps identify hiring patterns, seasonal trends, or sudden changes in job demand.
```

```python
# Horizontal Bar Chart: Top 10 Job Roles
plt.figure(figsize=(12, 6))
top_roles = df['Role'].value_counts().nlargest(10)
sns.barplot(y=top_roles.index, x=top_roles.values, palette="coolwarm")
plt.title("Top 10 Job Roles")
plt.xlabel("Number of Jobs")
plt.ylabel("Role")
plt.show()
```

```python
#key insights for Horizontal Bar Chart :
1.This horizontal bar chart highlights the top 10 most common job roles, showing which roles have the highest demand. 
2.It helps in understanding job market trends and the most sought-after positions.
```

```python
#Donut Chart: Work Type Distribution
plt.figure(figsize=(8, 8))
work_type_counts = df['Work Type'].value_counts()
plt.pie(work_type_counts, labels=work_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"), wedgeprops=dict(width=0.4))
plt.title("Work Type Distribution (Donut Chart)")
plt.show()
```

```python
#key insights for Donut Chart:
1.This donut chart shows the percentage distribution of different work types, making it easy to compare their proportions.
2.It helps identify the most and least common work types at a glance.
```

```python
#Time Series Heatmap: Job Posting Trends
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'])
df['Year-Month'] = df['Job Posting Date'].dt.to_period('M')

heatmap_data = df.groupby(['Year-Month', 'Work Type']).size().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt="d")
plt.title("Job Posting Trends Over Time")
plt.xlabel("Work Type")
plt.ylabel("Year-Month")
plt.show()
```

```python
#key insights for Time Series Heatmap:
1.This heatmap visualizes job posting trends over time across different work types, highlighting periods of high and low job availability. 
2.It helps identify seasonal hiring patterns and shifts in job demand.
```
